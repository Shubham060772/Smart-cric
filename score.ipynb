```python
import os
import pandas as pd
from tqdm import tqdm
from yaml import safe_load

# Folder path
folder_path = r'C:\Users\HP\data'

# Get list of valid files (ignoring .ipynb_checkpoints and directories)
filenames = [os.path.join(folder_path, file) for file in os.listdir(folder_path) 
             if not file.startswith('.ipynb_checkpoints') and os.path.isfile(os.path.join(folder_path, file))]

final_data = []  # Use a list instead of appending DataFrames directly
counter = 1

for file in tqdm(filenames, desc="Processing Files"):
    try:
        with open(file, 'r') as f:
            data = safe_load(f)  # Load YAML/JSON data
            
            if data:  # Check if data is not empty
                df = pd.json_normalize(data)  # Convert to DataFrame
                df['match_id'] = counter  # Add match_id
                final_data.append(df)  # Store DataFrame in list
                counter += 1
    except PermissionError:
        print(f"Skipping file due to permission error: {file}")
    except Exception as e:
        print(f"Error processing {file}: {e}")

# Create a single DataFrame after the loop
final_df = pd.concat(final_data, ignore_index=True) if final_data else pd.DataFrame()

# Display final DataFrame shape
print(f"Final DataFrame shape: {final_df.shape}")
```

```python
# List of desired base columns
import ast

base_columns = [
    'innings', 'meta.data_version', 'meta.created', 'meta.revision',
    'info.balls_per_over', 'info.city', 'info.competition', 'info.dates',
    'info.gender', 'info.match_type', 'info.outcome.by.runs',
    'info.outcome.winner', 'info.overs', 'info.player_of_match'
]

# Keep only valid columns
valid_base_columns = [col for col in base_columns if col in final_df.columns]

# Get player columns dynamically
player_columns = [col for col in final_df.columns if col.startswith('info.players.')]

# Combine columns to keep in dataframe
columns_to_keep = valid_base_columns + player_columns

# Filter dataframe
final_df = final_df[columns_to_keep]

# Now flatten players into player1, player2, ... columns for each row
def extract_players(row):
    all_players = []
    for col in player_columns:
        players = row[col]
        if isinstance(players, str):
            try:
                players = ast.literal_eval(players)  # convert string list to list
            except:
                pass
        if isinstance(players, list):
            all_players.extend(players)
    return all_players

# Extract players per row
final_df['all_players'] = final_df.apply(extract_players, axis=1)

# Find max number of players in any row (to know how many columns needed)
max_players = final_df['all_players'].apply(len).max()

# Create separate player columns
for i in range(max_players):
    final_df[f'player{i+1}'] = final_df['all_players'].apply(lambda x: x[i] if i < len(x) else None)

# Drop the temporary combined column if you want
final_df.drop(columns=['all_players'], inplace=True)

print(final_df.head())

```

```python
count = 1
delivery_df = pd.DataFrame()
for index, row in matches.iterrows():
    if count in [75,108,150,180,268,360,443,458,584,748,982,1052,1111,1226,1345]:
        count += 1
        continue
    count += 1
    ball_of_match = []
    batsman = []
    bowler = []
    runs = []
    player_of_dismissed = []
    teams = []
    batting_team = []
    match_id = []
    city = []
    venue = []
    for ball in row['innings'][0]['1st innings']['deliveries']:
        for key in ball.keys():
            match_id.append(count)
            batting_team.append(row['innings'][0]['1st innings']['team'])
            teams.append(row['info.teams'])
            ball_of_match.append(key)
            batsman.append(ball[key]['batsman'])
            bowler.append(ball[key]['bowler'])
            runs.append(ball[key]['runs']['total'])
            city.append(row['info.city'])
            venue.append(row['info.venue'])
            try:
                player_of_dismissed.append(ball[key]['wicket']['player_out'])
            except:
                player_of_dismissed.append('0')
    loop_df = pd.DataFrame({
        'match_id': match_id,
        'teams': teams,
        'batting_team': batting_team,
        'ball': ball_of_match,
        'batsman': batsman,
        'bowler': bowler,
        'runs': runs,
        'player_dismissed': player_of_dismissed,
        'city': city,
        'venue': venue
    })
    
    # **Use pd.concat() instead of append()**
    delivery_df = pd.concat([delivery_df, loop_df], ignore_index=True)
```

```python
# Convert 'runs_x' column to numeric to avoid KeyError
if 'runs_x' in output.columns:
    output['runs_x'] = pd.to_numeric(output['runs_x'], errors='coerce')

# Calculate cumulative runs per match
output['current_runs'] = output.groupby('match_id')['runs_x'].cumsum()

# Extract 'over' and 'ball_no' from 'ball'
output['over'] = output['ball'].astype(str).str.split(".").str[0].astype(int)
output['ball_no'] = output['ball'].apply(lambda x: int(str(x).split(".")[1]) if "." in str(x) else 0)

# Compute 'balls_bowled' safely
output['balls_bowled'] = output['over'] * 6 + output['ball_no']

import pandas as pd
import numpy as np

# Ensure 'player_dismissed' is treated as a string first
output['player_dismissed'] = output['player_dismissed'].astype(str)

# Replace '0' with 0, and all other values with 1
output['player_dismissed'] = output['player_dismissed'].apply(lambda x: 0 if x == '0' else 1)

# Convert to integer safely
output['player_dismissed'] = pd.to_numeric(output['player_dismissed'], errors='coerce').fillna(0).astype(int)

# Cumulative sum of wickets per match
output['player_dismissed'] = output.groupby('match_id')['player_dismissed'].cumsum()

# Calculate wickets left
output['wickets_left'] = 10 - output['player_dismissed']

# Print results
print(output[['match_id', 'player_dismissed', 'wickets_left']])


# Check if the column exists before calling .value_counts()
if 'wickets_left' in output.columns:
    print(output[output['wickets_left'] == 10].value_counts())
else:
    print("Column 'wickets_left' not found in the DataFrame.")
```

```python
import pandas as pd
from sklearn.model_selection import train_test_split
from sklearn.compose import ColumnTransformer
from sklearn.preprocessing import OneHotEncoder, StandardScaler
from sklearn.pipeline import Pipeline
from xgboost import XGBRegressor
from sklearn.metrics import r2_score, mean_absolute_error

# ✅ Assume final_df is already defined
# Define features and target
X = final_df.drop(columns=['match_id', 'runs_x', 'runs_y'])
y = final_df['runs_y']

# Train-test split
X_train, X_test, y_train, y_test = train_test_split(X, y, test_size=0.2, random_state=1)

# Column transformer to handle categorical columns
trf = ColumnTransformer([
    ('ohe', OneHotEncoder(sparse_output=False, drop='first'), ['batting_team', 'bowling_team', 'city'])
], remainder='passthrough')

# Pipeline with preprocessing + model
pipe = Pipeline(steps=[
    ('step1', trf),
    ('step2', StandardScaler()),
    ('step3', XGBRegressor(n_estimators=1000, learning_rate=0.2, max_depth=12, random_state=1))
])

# Fit the model
pipe.fit(X_train, y_train)

# Predict and evaluate
y_pred = pipe.predict(X_test)
print("R² Score:", r2_score(y_test, y_pred))
print("Mean Absolute Error:", mean_absolute_error(y_test, y_pred))
```
